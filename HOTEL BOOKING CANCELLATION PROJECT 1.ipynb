## Business Understanding

Dalam industri perhotelan, khususnya hotel, overbooking merupakan strategi yang umum digunakan untuk mengatasi pembatalan reservasi. Dengan berkembangnya platform online, pelanggan menjadi lebih mudah untuk melakukan pemesanan dan membatalkannya.

Pembatalan pemesanan dapat memberikan dampak yang signifikan terhadap operasional hotel. Meskipun pembatalan dapat membantu memaksimalkan pendapatan melalui strategi overbooking, hal ini juga dapat menimbulkan risiko terhadap reputasi hotel jika tidak dikelola dengan baik. Oleh karena itu, jika hotel dapat memprediksi pembatalan sejak dini, mereka dapat mengoptimalkan strategi overbooking dan meminimalkan potensi risiko.

## Introduction

Strategi overbooking akan bekerja paling optimal jika kita dapat memperkirakan berapa banyak pemesanan yang kemungkinan akan dibatalkan. Jika jumlah pembatalan diperkirakan terlalu rendah, maka banyak kamar yang bisa menjadi kosong dan tidak terpakai. Sebaliknya, jika jumlah pembatalan diperkirakan terlalu tinggi, hotel berisiko menerima terlalu banyak tamu tanpa ketersediaan kamar yang cukup, sehingga dapat menimbulkan ketidakpuasan pelanggan akibat overbooking.

Oleh karena itu, prediksi pembatalan yang akurat menjadi kunci untuk menjaga keseimbangan antara tingkat hunian kamar dan kepuasan tamu.

## Evaluation Metrics

Target (y):

**0 : Not Canceled**

**1 : Canceled**

**Confusion Matrix:**

| **Aktual / Prediksi**    | **Tidak Dibatalkan (0)** | **Dibatalkan (1)**  |
| ------------------------ | ------------------------ | ------------------- |
| **Not Canceled (0)** | True Negative (TN)       | False Positive (FP) |
| **Canceled (1)**       | False Negative (FN)      | True Positive (TP)  |


**False Negative (FN)**

Model memprediksi bahwa tamu akan datang, tetapi pada kenyataannya tamu tersebut membatalkan reservasi.

- Akibatnya, kamar menjadi kosong dan tidak terjual, padahal seharusnya dapat digunakan oleh tamu lain. Hal ini menyebabkan potensi pendapatan hotel menjadi hilang.

**False Positive (FP)**

Model memprediksi bahwa suatu reservasi akan dibatalkan, tetapi tamu tersebut justru datang.

- Kondisi ini menyebabkan hotel mengalami overbooking, sehingga berisiko menimbulkan ketidakpuasan pelanggan dan merusak reputasi hotel. Selain itu, tamu yang merasa dirugikan mungkin tidak akan kembali di masa mendatang.

**Kaitan dengan Strategi Overbooking dan Minimasi Biaya**

Dalam project ini, kesalahan prediksi seperti False Negative dan False Positive memiliki dampak langsung terhadap strategi overbooking dan efisiensi biaya operasional hotel.

- False Negative menyebabkan kehilangan peluang pendapatan akibat kamar kosong.
- False Positive berpotensi menimbulkan biaya kompensasi, keluhan pelanggan, serta penurunan loyalitas tamu.

Oleh karena itu, tujuan utama penelitian ini adalah meminimalkan kedua jenis kesalahan tersebut melalui pemilihan model prediksi yang optimal, sehingga hotel dapat menerapkan strategi overbooking secara lebih tepat dan efektif.

**Apakah Overbooking Merupakan Praktik yang Baik bagi Hotel? (Kelebihan dan Kekurangan Overbooking Hotel)**

Overbooking merupakan praktik yang umum dalam industri perhotelan untuk mengantisipasi pembatalan dan ketidakhadiran tamu pada saat terakhir. Namun, strategi ini memiliki kelebihan dan kekurangan.

**Kelebihan Overbooking Hotel**

- Mengurangi risiko pembatalan mendadak dan tamu yang tidak hadir
- Memaksimalkan tingkat hunian kamar
- Mencegah potensi kehilangan pendapatan
- Membantu hotel tetap kompetitif di platform Online Travel Agency (OTA)

**Kekurangan Overbooking Hotel**

- Hilangnya peluang pendapatan ketika tamu membatalkan dan kamar tidak terisi
- Ketidakpuasan tamu dan kerusakan reputasi hotel
- Ketidakefisienan dan gangguan dalam operasional hotel

**Dampak Kesalahan Prediksi**

Kedua jenis kesalahan prediksi dapat berdampak negatif terhadap pendapatan hotel:

- **False Negative (FN):** Model memprediksi bahwa tamu tidak akan membatalkan, tetapi tamu justru membatalkan. Hal ini menyebabkan kamar kosong dan kehilangan peluang pendapatan.
- **False Positive (FP):** Model memprediksi bahwa tamu akan membatalkan, tetapi tamu justru datang. Hal ini menyebabkan overbooking, yang dapat menimbulkan ketidaknyamanan bagi tamu dan merusak reputasi hotel.

Menurut pendapat saya, **overbooking (False Positive)** lebih merugikan dibandingkan kamar kosong (False Negative). Overbooking dapat menyebabkan kerugian jangka panjang karena tamu yang tidak puas cenderung tidak akan kembali. Selain itu, reputasi hotel juga akan terdampak, yang merupakan aset penting dalam industri perhotelan.

Meskipun ada pendapat bahwa overbooking lebih sedikit merugikan dibandingkan membiarkan kamar kosong, pada praktiknya overbooking sering kali memaksa hotel mengeluarkan biaya tambahan, seperti menyediakan transportasi atau memindahkan tamu ke hotel lain.

**Pemilihan Metrik Evaluasi**

Oleh karena itu, saya berpendapat bahwa kamar yang mengalami overbooking lebih merugikan dibandingkan kamar kosong, meskipun keduanya perlu diminimalkan. Untuk mengatasi hal ini, saya menggunakan **F2 Score** sebagai metrik evaluasi utama. F2 Score lebih menekankan pada recall dibandingkan precision, yang penting dalam konteks ini untuk mengurangi kasus pembatalan yang tidak terdeteksi.

Selain itu, saya juga menggunakan **ROC AUC Score** sebagai metrik evaluasi tambahan untuk menilai kinerja klasifikasi model secara keseluruhan.

## Modeling

### Train-Test Splitting

```python
X = df[num_features + cat_features]
y = df['is_canceled']
```

```python
# Split data: 70% train, 30% test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42)
```

### Feature Encoding & Scalling

```python
# Define preprocessors
cat_ohe = OneHotEncoder(handle_unknown='ignore')
num_scaler = RobustScaler()

# Full ColumnTransformer
preprocessor = ColumnTransformer([
    ('cat', cat_ohe, cat_features),
    ('num', num_scaler, num_features)
])
```

### Running Model (RobustScaled only)

```python
# Define custom F2 scorer
f2_scorer_scaled = make_scorer(fbeta_score, beta=2, average='binary')

models = {
    'logreg': LogisticRegression(max_iter=1000),
    'dt': DecisionTreeClassifier(),
    'rf': RandomForestClassifier(),
    'gboost': GradientBoostingClassifier(),
    'xgb': XGBClassifier(use_label_encoder=False, eval_metric='logloss')
}

# Create pipelines for each model
pipelines_scaled = {}
for name, model in models.items():
    pipelines_scaled[name] = ImbPipeline([
        ('preprocessor', preprocessor),
        ('model', model)
    ])
```

**Note:**
Dalam kasus ini, kami memprioritaskan model dengan nilai recall dan F2 score yang tinggi, karena kedua metrik tersebut menekankan kemampuan model dalam mendeteksi pembatalan secara efektif. Fokus ini sejalan dengan tujuan prediksi pembatalan reservasi hotel, di mana mengidentifikasi sebanyak mungkin pembatalan jauh lebih penting dibandingkan dengan ketepatan prediksi (precision).

```python
from sklearn.model_selection import cross_val_score

# Initialize empty list to store results
model_scores = []

# Loop through pipelines and calculate F2 score
for name, pipe in pipelines_scaled.items():
    scores = cross_val_score(pipe, X_train, y_train, cv=3, scoring=f2_scorer_scaled)
    avg_score = scores.mean()
    model_scores.append({'Model': name.upper(), 'F2 Score': avg_score})

# Convert to DataFrame
df_model_scores = pd.DataFrame(model_scores).sort_values(by='F2 Score', ascending=False)
```

```python
# Display the DataFrame
print(df_model_scores)
```

```python
top3_models = df_model_scores.head(3)
print(f"Top 3 Model (RobustScaled):\n\n{top3_models}")
```

### RobustScaled + SMOTE

```python
from imblearn.over_sampling import SMOTE

# Define custom F2 scorer
f2_scorer_sc_smote = make_scorer(fbeta_score, beta=2, average='binary')

# Define models
models = {
    'logreg': LogisticRegression(max_iter=1000),
    'dt': DecisionTreeClassifier(),
    'rf': RandomForestClassifier(),
    'gboost': GradientBoostingClassifier(),
    'xgb': XGBClassifier(use_label_encoder=False, eval_metric='logloss')
}

# Create pipelines for each model
pipelines_sc_smote = {}
for name, model in models.items():
    pipelines_sc_smote[name] = ImbPipeline([
        ('preprocessor', preprocessor),
        ('smote', SMOTE(random_state=42)),
        ('model', model)
    ])
```

```python
# Initialize empty list to store results
model_scores2 = []

# Loop through pipelines and calculate F2 score
for name, pipe in pipelines_sc_smote.items():
    scores = cross_val_score(pipe, X_train, y_train, cv=3, scoring=f2_scorer_sc_smote)
    avg_score = scores.mean()
    model_scores2.append({'Model': name.upper(), 'F2 Score': avg_score})

# Convert to DataFrame
df_model_scores2 = pd.DataFrame(model_scores2).sort_values(by='F2 Score', ascending=False)
```

```python
# Display the DataFrame
print(df_model_scores2)
```

```python
top3_models2 = df_model_scores2.head(3)
print(f"Top 3 Model (RobustScaled, SMOTE):\n\n{top3_models2}")
```

## Hyperparameter Tuning

### RobustScaled + SMOTE + Tuning (Random Forest)

```python
# Create pipeline with tuned params
rf_tuned = ImbPipeline([
    ('preprocessor', preprocessor),
    ('smote', SMOTE(random_state=42)),
    ('model', RandomForestClassifier(
        n_estimators= 100,
        min_samples_split= 5,
        min_samples_leaf= 2,
        max_depth= 20,
        criterion= 'gini',
        class_weight= None
    ))
])

# Fit to training data
rf_tuned.fit(X_train, y_train)

# Predict on test set
y_pred_rf = rf_tuned.predict(X_test)
```

```python
# Evaluate F2 score
f2_rf_tuned = fbeta_score(y_test, y_pred_rf, beta=2)
f2_rf_tuned
```

### RobustScaled + SMOTE + Tuning (XGBoost)

```python
# Create pipeline with tuned params
xgb_tuned = ImbPipeline([
    ('preprocessor', preprocessor),
    ('smote', SMOTE(random_state=42)),
    ('model', XGBClassifier(
        subsample= 0.6,
        n_estimators= 200,
        min_child_weight= 5,
        max_depth= 5,
        learning_rate= 0.1,
        gamma= 1,
        colsample_bytree= 1.0
    ))
])

# Fit to training data
xgb_tuned.fit(X_train, y_train)

# Predict on test set
y_pred_xgb = xgb_tuned.predict(X_test)
```

```python
# Evaluate F2 score
f2_xgb_tuned = fbeta_score(y_test, y_pred_xgb, beta=2)
f2_xgb_tuned
```

### RobustScaled + SMOTE + Tuning (Decission Tree)

```python
dt_tuned = ImbPipeline([
    ('preprocessor', preprocessor),
    ('smote', SMOTE(random_state=42)),
    ('model', DecisionTreeClassifier(
        max_depth=10,           
        min_samples_split=5,    
        min_samples_leaf=2,     
        criterion='gini',       
        random_state=42
    ))
])

# Fit ke data training
dt_tuned.fit(X_train, y_train)


# Predict ke test set
y_pred_dt = dt_tuned.predict(X_test)
```

```python
# Evaluate F2 score
f2_dt_tuned = fbeta_score(y_test, y_pred_dt, beta=2)
f2_dt_tuned
```
